```python
filtered_columns = [col for col in train_mnist.columns if not col.startswith('label_image')]
filtered_df = train_mnist[filtered_columns]
```

```python
filtered_df
```

```python
filtered_channel = [col for col in filtered_df.columns if col.startswith(('label', 'Oz'))]
channel_df = filtered_df[filtered_channel]
```

```python
channel_df
```

```python
Oz_label2 = channel_df[channel_df["label"] == 5].iloc[0]
```

```python
data = Oz_label2.drop("label")
```

```python
import plotly.graph_objs as go
import plotly.subplots as sp

color1 = dict(color='rgb(70, 130, 180)')
color2 = dict(color='rgb(0, 0, 139)')

# Sampling frequency
fs = 200  # Hz

# Create time array
time = jnp.arange(len(data)) / fs

# Assuming first_row_with_label_2 is your filtered row
fig = sp.make_subplots(rows=2, cols=1)

for col, value in data.items():
    fig.add_trace(go.Scatter(x=[col], y=[value], mode='markers', name=col), row=1, col=1)
fig.update_xaxes(title_text="Channels", row=1, col=1)
fig.update_yaxes(title_text="Amplitude", row=1, col=1)
    
fig.add_trace(go.Scatter(x=time, y=data.values, mode='markers', line=color2), row=2, col=1)
fig.update_xaxes(title_text="Time [s]", row=2, col=1)
fig.update_yaxes(title_text="Amplitude", row=2, col=1)

fig.update_traces(showlegend=False)
fig.update_layout(height=600, title=f'Data for Label {int(Oz_label2.iloc[0])}', template="plotly_white")
fig.show()
```
